```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
ipl=delivery.merge(match,left_on='match_id',right_on='id')
ipl.head()

```

```python
top100=ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(100).index.tolist()
new_ipl=ipl[ipl['batter'].isin(top100)]
```

```python
runs=new_ipl.groupby('batter')['batsman_runs'].sum()
balls=new_ipl.groupby('batter')['batsman_runs'].count()
strike_rate=(runs/balls)*100
strike_rate=strike_rate.reset_index()
strike_rate.head(50)
```

```python
out = ipl[ipl['player_dismissed'].isin(top100)]
number_of_out = out['player_dismissed'].value_counts()
avg = runs / number_of_out
avg = avg.reset_index()
avg.rename(columns={'index': 'batter', 0: 'avg'}, inplace=True)
avg = avg.merge(strike_rate, on='batter')     
number_of_out.head(20)'
```

```python
out = ipl[ipl['player_dismissed'].isin(top100)]
number_of_out = out['player_dismissed'].value_counts()
avg = runs / number_of_out
avg = avg.reset_index()
avg.rename(columns={'index': 'batter', 0: 'avg'}, inplace=True)
avg = avg.merge(strike_rate, on='batter')
number_of_out.head(20)
avg.head(30)
```

```python
def batsman_comp(*name):
    data = []
    for i in name:
        single = ipl[ipl['batter'] == i]
        performance = single.groupby('season')['batsman_runs'].sum().reset_index()
        trace = go.Scatter(x=performance['season'], y=performance['batsman_runs'],
                           mode='lines+markers', name=i)
        data.append(trace)

    layout = go.Layout(title='Batsman Record Comparator',
                       xaxis={'title': 'Season'},
                       yaxis={'title': 'Runs'})

    fig = go.Figure(data=data, layout=layout)
    pyo.plot(fig, filename='year_by_year')
```

```python
batsman_comp('AB de Villiers','V Kohli','MS Dhoni','SK Raina')
```

```python
batsman_comp('AB de Villiers','V Kohli','MS Dhoni','MK Pandey' ,'S Dhawan','RG Sharma','SK Raina ')
```

```python
iw=top10_df.groupby(['batter','inning'])['batsman_runs'].sum().reset_index()
mask1=iw['inning']==1
mask2=iw['inning']==2
one=iw[mask1]
two=iw[mask2]
mask1=iw['inning']==1
one.rename(columns={'batsman_runs':'1st Innings'},inplace=True)
two.rename(columns={'batsman_runs':'2nd Innings'},inplace=True)
final=one.merge(two,on='batter')[['batter','1st Innings','2nd Innings']]
final.head(15)
```

Overlayed Bar Graph

```python
trace1=go.Bar(x=final['batter'],y=final['1st Innings'],name='1st innings',marker={'color':'#00a65a'})
trace2=go.Bar(x=final['batter'],y=final['2nd Innings'],name='2nd innings')
data=[trace1,trace2]
layout=go.Layout(title='Inning wise Scores',xaxis={'title':'Batsman'},yaxis={'title':'Runs'},barmode='overlay')
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# 8.Histogram

```python
# this batter who play atlist 150 ball
x=delivery.groupby('batter')['batsman_runs'].count()>150
x=x[x].index.tolist()
new=delivery[delivery['batter'].isin(x)]
runs=new.groupby('batter')['batsman_runs'].sum()
balls=new.groupby('batter')['batsman_runs'].count()
strick_rate=(runs/balls)*100
strick_rate=strick_rate.reset_index()
strick_rate
```

```python
# Plot Histogram
trace=go.Histogram(x=strick_rate['batsman_runs'])
data=[trace]
layout=go.Layout(title='Strick Rate Analysis',xaxis={'title':'Strike Rates'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
# Plot Histogram
trace=go.Histogram(x=strick_rate['batsman_runs'],xbins={'size':2})
data=[trace]
layout=go.Layout(title='Strick Rate Analysis',xaxis={'title':'Strike Rates'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
```
